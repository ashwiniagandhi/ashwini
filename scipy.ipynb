```python
#gaussian distribution with 1_d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
r=np.random.randn(1000)
plt.hist(r,bins=40)
```

```python
#sampling from gaussian distribution sperical means it is corelated and independent with each other#variance is same #
r=np.random.randn(1000,2)#data sred out in circle using scatter plot
plt.scatter(r[:,0],r[:,1])
```

```python
# if u want elliptical gaussian.variance different for each dimension.std 5 and mean is 2
r[:,1]=5*r[:,1]+2
plt.scatter(r[:,0],r[:,1])
#first row r[:,1]
```

```python
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')#ecllipse.dimension corelated with each other
plt.show()
```

```python
#sampling from genral mutivariant normal
cov=np.array([[1,0.5],[0.5,4]])#define covariance matrix.variance 1 is firstdimension and variance 3 is second dimension and covariance between two dimaension iso.5
from scipy.stats import multivariate_normal as mvn 
```

```python
m=np.array([0,2]) #third para is size its specify number ofsample u want n first two para is mean and varince
r=mvn.rvs(mean=m,cov=cov,size=1000)
#create randomvariable sample
```

```python
plt.scatter(r[:,0],r[:,1])
```

```python
plt.scatter(r[:,0],r[:,1])

plt.axis('equal')
```

```python
import numpy as np#by using numpy array
r=np.random.multivariate_normal(mean=m,cov=cov,size=1000)
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')
plt.show()
```
