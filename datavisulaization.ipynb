```python
import pandas as pd  
import matplotlib.pyplot as plt  
import seaborn as sns 
```

```python
sns.set_style("darkgrid")  
sns.FacetGrid(iris, hue="location", size=4).map(plt.scatter, "temp", "humid").add_legend()  
plt.title('pune city')  
plt.xlabel('X-Label')  
plt.ylabel('Y-Label')  
plt.show() 
```

```python
sns.boxplot(x='temp',y='humid',data=iris)
plt.show()
```

```python
sns.distplot( iris["humid"], bins=20 )
plt.show()
```

```python
sns.countplot('humid', data=iris)
plt.show()
```

```python
sns.violinplot(x='temp',y='humid',data=iris)
plt.show()
```

```python
iris.plot(kind='scatter', x='humid', y='temp',label='iris',color='red')
plt.show() 
```

```python
sns.FacetGrid(iris, hue="location", size=5).map(plt.scatter, "temp", "humid").add_legend()
plt.show() 
```

```python
sns.pairplot(iris,hue='location',kind='reg')
plt.show() 
```

```python
from pandas.plotting import andrews_curves    
andrews_curves(iris.drop("date",axis=1),'location') 
plt.show() 
```

```python
sns.heatmap(iris.corr(),linewidth=0.3,vmax=1.0,square=True, linecolor='black',annot=True)
plt.show() 
```

```python
sns.jointplot(x='temp',y='humid', data=iris, kind='resid')
plt.show() 
```
