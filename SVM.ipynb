```python
from sklearn.model_selection import train_test_split
```

```python
# in Df dataset 1st 4 features used for training model,and last two target col then remove
x=df.drop(['target','flower_name'],axis='columns')
x
```

```python
y=df.target
y
```

```python
train_test_split(x,y,test_size=0.2)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
len(x_train),len(y_train),len(x_test),len(y_test)
```

```python
#svc is classsifier and  create svm classifier
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python

#cids by default 1.0,increasing regulazition to decrease score
from sklearn.svm import SVC
model=SVC(C=100)
```

```python
model.fit(x_train, y_train)
model.score(x_test,y_test)
```

```python
#decresing score if gamma is incresing
model=SVC(gamma=100)
model.fit(x_train, y_train)
model.score(x_test,y_test)
```
