```python
#tensorflow>2.0
from tensorflow.keras.preprocessing.text import one_hot#use for converting sentence into one hot encoder
```

```python
##sentences
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'i am a good developer',
     'your video are good',
     'understand the meaning of words',
     'i am good girl']
```

```python
sent
```

```python

voc_size=10000#define dictonary size,this is vocabulary size
```

```python
#one hot  representation
onehot_rep=[one_hot(words,voc_size)for words in sent]#i pass sentence which are presented word and to pass the one_hot function 
print(onehot_rep)#convert each sentence into index. see 1st and 2nd sentence are similar but some word are different and this index size also change and which word are same having this word index size also same
#only last index of 1st two sentence are changed.
```

```python
#word embedding representation for creating embedding matrix
#feature representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences#use for size of sentence
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8#length of sentence size are fixed
embedded_docs=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)#padding as a pre means add 0 in the front /starting point .if padding as a post then zeros add ending
print(embedded_docs)#try to compelte 8 word sentence sent_length is 8.
```

```python
dim=15#dimension intialized means features
```

```python
model=Sequential()#intialized sequential layer
model.add(Embedding(voc_size,10,input_length=sent_length))#addd embedding layer best on demsion featurized representation
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs)[0])#1 st sentence all word are convert into dimension of 10 vectors.1st 'o' are convert into dimension of 10 vector
#1st 4 word line are same bcoz 1st 4 word are zero
```
