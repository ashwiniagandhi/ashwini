```python
from matplotlib import pyplot as pt
```

```python
age_x=[23,54,97,98]
```

```python
dev_y=[366,425,433,489]
```

```python
pt.title('salary by age')
pt.xlabel('age')
pt.ylabel('salary')
pt.plot(age_x,dev_y)
pt.show()
```

```python
py_dev_y=[344,654,765,865]
```

```python
pt.title('salary by age')
pt.xlabel('age')
pt.ylabel('salary')
pt.plot(age_x,dev_y,label='all devs')
pt.plot(age_x,py_dev_y,label='python')
pt.legend(['all devs','python'])
pt.show()
```

```python
pt.title('salary by age')
pt.xlabel('age')
pt.ylabel('salary')
pt.plot(age_x,dev_y,'k--',label='all devs')
pt.plot(age_x,py_dev_y,'r',label='python')
pt.legend(['all devs','python'])
pt.show()
```

```python
pt.title('salary by age')
pt.xlabel('age')
pt.ylabel('salary')
pt.plot(age_x,dev_y,color='k',linestyle='--',marker='.',label='all devs')
pt.plot(age_x,py_dev_y,color='r',marker='o',label='python')
pt.legend(['all devs','python'])
pt.show()
```

```python
pt.title('salary by age')
pt.xlabel('age')
pt.ylabel('salary')
pt.plot(age_x,dev_y,color='#444444',linestyle='--',marker='.',label='all devs')
pt.plot(age_x,py_dev_y,color='r',marker='o',label='python')
pt.legend(['all devs','python'])
```

```python
js_dev_y=[200,300,500,600]
pt.title('salary by age')
pt.xlabel('age')
pt.ylabel('salary')
pt.plot(age_x,dev_y,color='b',linestyle='--',marker='.',label='all devs')
pt.plot(age_x,js_dev_y,color='#5a7d9a',linewidth=7,marker='o',label='python')
pt.plot(age_x,py_dev_y,color='#5a7d9a',marker='o',label='javascript')
pt.legend(['all devs','python','javascript'])
```

```python
js_dev_y=[200,300,500,600]
pt.title('salary by age')
pt.xlabel('age')
pt.ylabel('salary')
pt.plot(age_x,dev_y,color='b',linestyle='--',marker='.',label='all devs')

pt.plot(age_x,py_dev_y,color='#5a7d9a',marker='o',label='javascript')
pt.plot(age_x,js_dev_y,color='#5a7d9a',linewidth=7,marker='o',label='python')
pt.legend()
pt.grid()
pt.tight_layout()
```
