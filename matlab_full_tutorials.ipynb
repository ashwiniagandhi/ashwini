```python
import matplotlib.pyplot as plt
import numpy as nm
%matplotlib inline
```

```python
company=['gooogle','yahoo','Tcs','Apple']
revenue=[134,145,165,123]
profit=[43,65,75,65]
```

```python
a=nm.arange(len(company))
a
```

```python
plt.bar(a,revenue)
```

```python
plt.xticks(a,company)
plt.bar(a,revenue)
```

```python
plt.xticks(a,company)
plt.xlabel('company name')
plt.ylabel('revenue')
plt.title('companies revenue')
plt.bar(a,revenue)
```

```python
plt.xticks(a,company)
plt.xlabel('company name')
plt.ylabel('revenue')
plt.title('companies revenue')
plt.bar(a,revenue,label='Revenues')
plt.legend()
```

```python

plt.xticks(a,company)
plt.xlabel('company name')
plt.ylabel('revenue')
plt.title('companies revenue')
plt.bar(a,revenue,label='Revenues')
plt.bar(a,profit,label='profit')
plt.legend()
```

```python

plt.xticks(a,company)
plt.xlabel('company name')
plt.ylabel('revenue')
plt.title('companies revenue')
plt.bar(a-0.4,revenue,label='Revenues')
plt.bar(a+0.3,profit,label='profit')
plt.legend()
```

```python

plt.xticks(a,company)
plt.xlabel('company name')
plt.ylabel('revenue')
plt.title('companies revenue')
plt.bar(a,revenue,width=0.4,label='Revenues')
plt.bar(a,profit,width=0.7,label='profit')
plt.legend()
```

```python

plt.yticks(a,company)
plt.xlabel('company name')
plt.ylabel('revenue')
plt.title('companies revenue')
plt.barh(a,revenue,label='Revenues')
plt.barh(a,profit,label='profit')
plt.legend()
```
